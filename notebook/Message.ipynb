```python
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

```

```python
llm = ChatGoogleGenerativeAI(model="gemini-3-flash-preview", convert_system_message_to_human=True)

messages = [
    SystemMessage("You are a helpful Gen AI tutorial assistant."),  
    HumanMessage("Best Model for Gen AI? Only in two words")
]

response = llm.invoke(messages)

print(response)
```
