# ReAct Agent Demo with Custom Tools (LangChain)

This notebook demonstrates a **simple ReAct (Reason + Act) agent** using LangChain.

We will create **three custom tools**:
1. IceCream tool
2. States tool
3. Schools tool

The agent will:
- Think step‑by‑step
- Choose the correct tool
- Use tool output
- Produce the final answer

`verbose=True` enables **ReAct thinking logs**.

## Import Libraries

```python
from langchain.agents import initialize_agent, AgentType
from langchain.tools import Tool
```

## Create Dummy Tools

```python
def icecream_tool(query: str):
    data = {
        "chocolate": "Chocolate ice cream costs $5",
        "vanilla": "Vanilla ice cream costs $4",
        "strawberry": "Strawberry ice cream costs $6"
    }
    for key in data:
        if key in query.lower():
            return data[key]
    return "Ice cream not found"
```

```python
def states_tool(query: str):
    data = {
        "telangana": "Capital of Telangana is Hyderabad",
        "karnataka": "Capital of Karnataka is Bangalore",
        "tamil nadu": "Capital of Tamil Nadu is Chennai"
    }
    for key in data:
        if key in query.lower():
            return data[key]
    return "State not found"
```

```python
def schools_tool(query: str):
    data = {
        "oakridge": "Oakridge is an international school in Hyderabad",
        "silver oaks": "Silver Oaks is an IB school",
        "dps": "Delhi Public School has many branches in India"
    }
    for key in data:
        if key in query.lower():
            return data[key]
    return "School not found"
```

## Register Tools

```python
tools = [
    Tool(
        name="IceCreamTool",
        func=icecream_tool,
        description="Use this tool to get information about ice cream flavors and prices"
    ),
    Tool(
        name="StatesTool",
        func=states_tool,
        description="Use this tool to get information about Indian states and their capitals"
    ),
    Tool(
        name="SchoolsTool",
        func=schools_tool,
        description="Use this tool to get information about schools"
    )
]
```

## Initialize LLM

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-3-flash-preview", convert_system_message_to_human=True)
```

## Create ReAct Agent

```python
agent = initialize_agent(
    tools,
    llm,
    verbose=True
)

agent.run("What is the capital of Telangana and how much does chocolate ice cream cost?")
```

## Ask Questions

The agent will **think → choose tool → observe → answer**.

```python
agent.run("What is the capital of Delhi?")
```

```python
agent.run("Tell me about Silver Oaks school")
```

```python
agent.run("What is the price of chocolate ice cream?")
```
